```python
# initial values
a = [0.5, 0.5, 2.5, -1.5, -0.5]

# dataset
from sklearn import datasets 
iris = datasets.load_iris()

features = iris.data
labels = iris.target

xt = []
store = [list(val) for val in features]
for lis in store:
    xt.append(tuple(lis))

t = []
for val in labels:
    if val == 0:
        t.append(1)
    elif val == 1:
        t.append(-1)
    elif val == 2:
        t.append(-1)
        
Y = []
store = [val for val in features]
for i in range(len(labels)):
    if t[i] == -1:
        store[i] = -store[i]

new_store = [list(val) for val in store]

for i in range(len(labels)):
    new_store[i].insert(0, t[i])

for lis in new_store:
    Y.append(tuple(lis))
    
# margin vector
b = [1]* len(Y)

# learning rate 
n = 0.01

# iterations 
iterations = 300

result = []
for o in range(int(iterations/ len(Y))):
    for i in range(len(Y)):
        a_prev = a
        y = Y[i]
        
        
        #calculate ay
        ay = np.dot(a,y)
        
        # calculate update part 
        update = np.zeros(len(y))
        for j in range(len(y)):
            update[j] = n* (b[i] - ay) * y[j]
            
        # add update part to a 
        a = np.add(a, update)
        
        # append result 
        result.append((str(i+1+(len(Y) * o)), np.round(a_prev,4), np.round(y,4), np.round(ay,4), np.round(a,4)))
        
# pretty table
pt = PrettyTable(('iteration', 'a', 'y', 'ay', 'a_new'))
for row in result: pt.add_row(row)

pt.align['iteration'] = 'c'
pt.align['a'] = 'l'
pt.align['y'] = 'l'
pt.align['ay'] = 'l'
pt.align['a_new'] = 'l'

print(pt)
```
